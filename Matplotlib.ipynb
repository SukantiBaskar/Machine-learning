```python
#importing matplot lib
import matplotlib.pyplot as plt
```

```python
#importing numpy to get the data
import numpy as np
```

```python
x=np.linspace(0,10,100)#100 evenly spaces values between 0 and 10
y=np.sin(x)
z=np.cos(x)
print(x)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title,x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
#bar chart
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','Latin','French','Spanish','German']
people=[100,50,120,70,40]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#pie chart
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','Latin','French','Spanish','German']
people=[100,50,120,70,40]
ax.pie(people, labels=languages,autopct='%1.3f%%')
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='b')
ax.scatter(x,z,color='r')
plt.show()
```

```python
#3D scatter plot
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
