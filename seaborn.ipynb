```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```
